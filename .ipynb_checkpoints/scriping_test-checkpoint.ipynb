<h1>スクレイピング + データ可視化</h1>
<div font-size=30>スクレイピングサイト :<a href="https://books.toscrape.com/">Books to Scrape</a>

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import japanize_matplotlib
```

<h2>スクレイピング</h2>

```python
#url指定
base_url = "https://books.toscrape.com/"
```

```python
#ページ取得
res = requests.get(base_url)
```

```python
#httpエラー例外処理
try:
    res.raise_for_status()
except requests.exceptions.HTTPError as e:
    print(e)
```

```python

soup = BeautifulSoup(res.text, "html.parser")
```

```python
categories = []

#カテゴリとリンク取得
for a in soup.select("ul.nav-list li a"):
    #カテゴリ名
    category = a.get_text(strip=True)
    #各カテゴリのurl
    link = base_url + a["href"]
    if category != "Books":
        categories.append((category, link))

data = []

# 各カテゴリページにアクセス
for category, url in tqdm(categories):
    res = requests.get(url)
    res.raise_for_status()
    cat_soup = BeautifulSoup(res.text, "html.parser")

    # 商品一覧を抽出
    for book in cat_soup.select("article.product_pod"):
        title = book.h3.a["title"]
        price = book.select_one(".price_color").get_text(strip=True)
        data.append([category, title, price])
```

<h2>csv化・集計</h2>

```python
#データフレーム型に変換
df = pd.DataFrame(data, columns=["category", "title", "price(£)"])
```

```python
#無駄な文字削除
df["price(£)"] = df["price(£)"].str.replace("[£Â]", "", regex=True).astype(float)
```

```python
#csv保存
df.to_csv("./csv/books_price_data.csv",index=False, encoding="utf-8-sig")
```

```python
#カテゴリごとの値段の平均・中央値取得
summary = df.groupby("category")["price(£)"].agg(["mean", "median"]).reset_index()
#csv保存
summary.to_csv("./csv/books_price_ave_med_data.csv",index=False, encoding="utf-8-sig")
```

```python
#集計値可視化・グラフ保存
summary.plot(x="category", y = ["mean", "median"],kind="bar",figsize=(10,5))
plt.title("カテゴリごとの本の値段 平均値と中央値")
plt.ylabel("category")
plt.ylabel("Price (£)")
plt.legend(["Mean", "Median"])
plt.tight_layout()
plt.savefig("./img/books_ave_med_data.jpg")
plt.show()
```
